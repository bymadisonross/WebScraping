```python
import requests
# library used for HTTP requests
```

```python
r = requests.get('https://www.bonappetit.com/recipe/lemon-saffron-salmon-with-dill-rice')
# import recipe URL
```

```python
print(r.text)
# the output text of the page (r) in HTML
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(r.text, 'html.parser')
# import library Beautiful Soup which lets you parse HTML
```

```python
scripts = soup.find_all('script', {'type':'application/ld+json'})
# makes a list of the script tags with the attribute type = application/ld+json
```

```python
for script in scripts:
    print(script.text)
```

```python
import json
```

```python
for script in scripts:
    raw_json = json.loads(script.text)
    json_type = raw_json['@type']
    if raw_json['@type'] == 'Recipe':
        recipe = raw_json
        
# assigns the variable where the @type is Recipe
```

```python
print(recipe)
```
